```python
import pandas as pd
import numpy as np
```

```python
PostalCodes = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
PostalCodes=PostalCodes[0]
PostalCodes
```

```python
PostalCodes.drop(PostalCodes.index[PostalCodes['Borough'] == 'Not assigned'], inplace = True)
```

```python
X=PostalCodes
```

```python
X=X.groupby('Postcode')['Neighbourhood'].apply(','.join).reset_index()
X
```

```python
PostalCodes=pd.merge(X,PostalCodes,on='Postcode')
PostalCodes
```

```python
PostalCodes.drop(['Neighbourhood_y'],axis=1,inplace=True)
PostalCodes
```

```python
PostalCodes=PostalCodes[['Postcode','Borough','Neighbourhood_x']]
PostalCodes
```

```python
PostalCodes=PostalCodes.drop_duplicates(subset='Postcode').reset_index()
PostalCodes
```

```python
PostalCodes.drop(columns=['index'],inplace=True)
PostalCodes.rename(columns={'Neighbourhood_x' : 'Neighbourhood'},inplace=True)
PostalCodes
```
