## Part-1( Reading From Wikipedia and Generating Necessary DataFrame ) 

## Importing Libraries

```python
import pandas as pd
import numpy as np
```

### Reading The Postal Code Values From Wikipedia Page

```python
PostalCodes = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

### Raw Postal Code Values

```python
PostalCodes=PostalCodes[0]
PostalCodes
```

### Removing "Not Assigned" Values From DataFrame

```python
PostalCodes.drop(PostalCodes.index[PostalCodes['Borough'] == 'Not assigned'], inplace = True)
```

### Joining Neighbourhood Values Separated by Commas & Merging with Postal Code Dataframe

```python
X=PostalCodes
X=X.groupby('Postcode')['Neighbourhood'].apply(','.join).reset_index()
PostalCodes=pd.merge(X,PostalCodes,on='Postcode')
PostalCodes.drop(['Neighbourhood_y'],axis=1,inplace=True)
```

### ReDesigning and Removing Duplicate Values Along with Reseting the Index Values

```python
PostalCodes=PostalCodes[['Postcode','Borough','Neighbourhood_x']]
PostalCodes=PostalCodes.drop_duplicates(subset='Postcode').reset_index()
PostalCodes.drop(columns=['index'],inplace=True)
PostalCodes.rename(columns={'Neighbourhood_x' : 'Neighbourhood'},inplace=True)
PostalCodes
```

### Removing "Not Assigned" From Neighbourhood and Replacing with Name of the Borough

```python
PostalCodes.loc[PostalCodes['Neighbourhood']=="Not assigned",'Neighbourhood']=PostalCodes.loc[PostalCodes['Neighbourhood']=="Not assigned",'Borough']
```

### Generating a Function for Cleaning the Data

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

## Exploring the Neighbourhoods in Toronto

### Function to get all the Neighbourhoods in Toronto

```python

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### Onehot Encoding

```python
toronto_onehot = pd.get_dummies(Toro_neighvenues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehot['Neighborhood'] = Toro_neighvenues['Neighborhood'] 
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]
toronto_onehot.head()
```

```python
toronto_group = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_group
```

### Top 5 Venues

```python
num_top_venues = 5

for hood in toronto_group['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_group[toronto_group['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

### Putting into DataFrame

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

### Top 5 Venues in a DataFrame

```python
num_top_venues = 5
indicators = ['st', 'nd', 'rd']
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_group['Neighborhood']

for ind in np.arange(toronto_group.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_group.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

### Using KMeans to Generate a Model

```python
kclusters = 5
toronto_grouped_clustering = toronto_group.drop('Neighborhood', 1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)
kmeans.labels_[0:5]
```

```python
toronto_merg = Toro_Code
toronto_merg['Cluster Labels'] = kmeans.labels_
toronto_merg = toronto_merg.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighbourhood')
toronto_merg.head()
```

### Map Clusters

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=13)
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merg['Latitude'], toronto_merg['Longitude'], toronto_merg['Neighbourhood'], toronto_merg['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=10,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
